```python
import pandas as pd
import sqlite3
import json
import os

# -----------------------------
# Step 1: Load Orders (CSV)
# -----------------------------
orders = pd.read_csv("orders.csv")

# -----------------------------
# Step 2: Load Users (JSON)
# -----------------------------
with open("users.json", "r") as f:
    users_data = json.load(f)

users = pd.DataFrame(users_data)

# -----------------------------
# Step 3: Load Restaurants (SQL)
# -----------------------------
# Create in-memory SQLite database
conn = sqlite3.connect(":memory:")

# Execute SQL file
with open("restaurants.sql", "r") as f:
    sql_script = f.read()

conn.executescript(sql_script)

# Read restaurants table into pandas
restaurants = pd.read_sql_query(
    "SELECT * FROM restaurants", conn
)

# -----------------------------
# Step 4: LEFT JOIN the datasets
# -----------------------------
# orders â†’ users
orders_users = pd.merge(
    orders,
    users,
    how="left",
    on="user_id"
)

# orders_users â†’ restaurants
final_df = pd.merge(
    orders_users,
    restaurants,
    how="left",
    on="restaurant_id"
)

# -----------------------------
# Step 5: Save Final Dataset
# -----------------------------
output_file = "final_food_delivery_dataset.csv"
final_df.to_csv(output_file, index=False)

print("âœ… Final dataset created successfully!")
print(f"ğŸ“ File name: {output_file}")
print("ğŸ“Š Dataset shape:", final_df.shape)
print(f"The dataset is saved at: {os.getcwd()}/{output_file}")
```
